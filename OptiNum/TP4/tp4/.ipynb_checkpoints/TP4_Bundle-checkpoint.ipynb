## Affectation linéaire
---

Considérons le problème d'affectation (P):
$$
\begin{aligned}
\min \quad \sum_{i=1}^n\sum_{j=1}^na_{ij}u_{ij}\\
\quad \sum_{j=1}^n\;u_{ij} = 1,\quad \textrm{pour tout }i=1,\dots,n,\\
\quad \sum_{i=1}^n\;u_{ij} = 1,\quad \textrm{pour tout }j=1,\dots,n,\\
\quad u_{ij}\in \{0,1\},\quad \textrm{pour tout }i,j=1,\dots,n.
\end{aligned}
$$


1. Écrire (P) sous forme matricielle, et donner une interpretation matricielle du probèlme. 
>$$
\begin{aligned}
\textrm{min}_U \ \langle A, U \rangle \\
U \ \textrm{matrice de permutations}
\end{aligned}
$$
1. On veut dualiser les $n$ premieres contraintes dans (P), écrire le lagrangien $L(\cdot;\cdot)$ associe a ce problème.
>On reécrit le problème comme une maximisation: 
>$$
\begin{aligned}
-\textrm{max}_U \ -\langle A, U \rangle \\
U \ \textrm{matrice de permutations}
\end{aligned}
$$
>Le lagrangien associé est:
>$$
\begin{aligned}
L(U, \lambda) &= \sum_{i=1}^n\sum_{j=1}^n - a_{ij}u_{ij} + \sum_{i=1}^n - \lambda_i \Big(\sum_{j=1}^nu_{ij} - 1\Big)\\
              &= \sum_{j=1}^n\sum_{i=1}^n u_{ij}(- a_{ij}- \lambda_i) + (1\dots 1)^T\lambda 
\end{aligned}
$$
1. Definir la fonction duale associée.
>$$
\begin{aligned}
\theta(\lambda) = \underset{\begin{aligned}\forall j \ \sum_{i=1}^n u_{ij} = 1 \\ \forall ij \ u_{ij} \in \{0, 1\}\end{aligned}}{\textrm{max}_U} L(U, \lambda) 
\end{aligned}
$$   
>Il s'agit d'un problème séparable en $j$:
>$$
\begin{aligned}
\theta(\lambda) &= (1\dots 1)^T\lambda + \sum_{j=1}^n \underset{\begin{aligned}\sum_{i=1}^n u_{ij} = 1 \\ \forall i,j \ u_{ij} \in \{0, 1\}\end{aligned}}{\textrm{max}_U} \sum_{i=1}^n u_{ij}(- a_{ij}- \lambda_i) \\
&= (1\dots 1)^T\lambda + \sum_{j=1}^n \textrm{max}_i (- a_{ij}- \lambda_i)
\end{aligned}
$$ 
1. Montrer qu'on peut calculer explicitement $U_{\lambda}$.
>On peut calculer $U_{\lambda}$ comme suit :
$$
\begin{aligned}
\forall j \ \ U_{\lambda, \ i} =  \textrm{argmax}_i (- a_{ij}- \lambda_i)
\end{aligned}
$$ 

## Simulateur de la fonction duale
---

Pour la suite du TP, on fixe les donnees : 
on suppose qu'on est en dimension $n=10$ 
et que les couts sont donnes par la matrice 
$$                                   A = [1:n]*[1:n]^\top $$

> * Calculer la matrice $U_{\lambda}$ 
> * Utiliser cette matrice $U_{\lambda}$ 
pour ecrire un simulateur ``simdual`` de la fonction duale $\theta$ dont les sorties sont, pour l'entree ``lambda``:
  - $\theta (\lambda) = L(U_{\lambda},\lambda)$, la valeur de la fonction duale,
  - $g(\lambda) = -c(U_{\lambda})$ (avec les notations du cours). 
      
> *Indice: le vecteur $-c(U_{\lambda})$ est un sous-gradient de $\theta$ .*


*Remarque : Mettre uniquement $\lambda$ en parametre d'entree du
simulateur pour bien garder la forme des TPs
precedents. Rigoureusement, la matrice $A$ devrait aussi etre en
parametre du simulateur, mais pour faire simple, on la considere
comme une variable globale.*


```python
def theta(lambda_, A):
    n = A.shape[0]
    theta = float(sum(lambda_))
    for j in range(n):
        theta += max(-(A + lambda_)[:,j]) + sum(lambda_)
    return theta

def U_(lambda_, A):
    n = A.shape[0]
    U = np.zeros((n, n))
    for j in range(n):
        i = np.argmax(-(A + lambda_)[:,j])
        U[i][j] = 1
    return U

def subgrad(U, A):
    n = A.shape[0]
    sg = np.zeros((n, 1))
    for i in range(n):
        sg[i] = (1 - sum(U[i,:]))
    return sg

def simdual(lambda_, A):
    return theta(lambda_, A), subgrad(U_(lambda_, A), A) 
```

## Résolution du problème dual
---

On définit la fonction `minimize` qui minimise une fonction étant donné un simulateur de cette fonction et une fonction donnant $x_{k+1}$ à partir des données de l'iteration `k`.

```python
def minimize(f_simulator, next_x, x0, nb_iter):
    """
    Minimizes the function simulated by 'f_simulator'. 
    
    Parameters:
    - the simulator 'f_simulator' of a function f
    - a function 'next_x' that can computes x_k+1 based on data computed at the iteration k
    - the number of iterations 'nb_iter'
    - the initial value of x 'x0'
    
    Returns:
    - the final point x that aproaches the optimal solution of minimization problem
    - the list of x_k found during the 'nb_iter' iterations
    - the list of corresponding values of f
    - the list of corresponding values of subradient of f
    """
    
    # Initialisation
    x           = np.copy(x0)
    x_tab       = [np.copy(x0)]
    f_tab       = [f_simulator(x)[0]]
    subgrad_tab = [f_simulator(x)[1]]
    
    for k in range(nb_iter):
        # Simulation of f and its subgradient in x 
        f, subgrad = f_simulator(x)
        
        # Updating the list of xs and the corresponding values of f and its subgradient
        f_tab       += [f]
        subgrad_tab += [np.copy(subgrad)]
        x            = next_x(x, x_tab, f_tab, subgrad_tab)
        x_tab       += [np.copy(x)]
    
    return x, x_tab, f_tab, subgrad_tab
```

### Résolution par sous-gradient

> Trouver un minimum de la fonction duale par la méthode du sous-gradient (c.f. cours sur l'optimisation non-lisse)

```python
def step(k):
    """
    Step for the subgradient algorithm. 
    This step slows down the descent when the number of iterations increases.
    """
    return 1/(1 + (k/1000)**0.5)


def subgrad_descent_next_x(x, x_tab, f_tab, subgrad_tab):
    k = len(x_tab)
    return x - step(k)*subgrad_tab[-1]
```

### Résolution par algorithme des faisceaux (Bundle)

> Trouver un minimum de la fonction duale par la méthode des faisceaux en formulant chaque mise à jour comme un problème QP et en faisant appel à cvxopt comme dans le TP LP/QP.

####  Rappel de cours sur la méthode des faisceaux :

Soit $\theta : R^n\rightarrow R,\lambda\mapsto\theta(\lambda)$ une fonction convexe 
(non differentiable a priori). 

Rappelons rapidement le principe de la methode 
des faisceaux pour minimiser $\theta$.

A l'iteration $k$, on dispose du  faisceau d'information
$$
\{(\theta(\lambda^j),g^j):j=1,\ldots,k\}.
$$
(Dans tout ce qui suit, les indices superieurs noteront un numero
d'iteration.)



On definit $\hat\lambda^k$ comme le meilleur des
$\lambda^j$, i.e.
$\theta(\hat\lambda^k)\leqslant\theta(\lambda^j)$, $j=1,\ldots,k$.
L'itere suivant est alors obtenu comme solution de :
$$ 
  \begin{aligned}
  \min r+\frac{1}{2} \| \lambda-\hat\lambda^k \|^2 \\
  (\lambda,r)\in R^n\times R \\
  \quad r\geqslant \theta(\lambda^j)+(\lambda-\lambda^j)^\top g^j,  j=1,\ldots,k
  \end{aligned}
$$

#### Matrices pour le solver `qp` de cvxopt :

![](ressources/1.png)

```python
def P_(lambda_):
    n = lambda_.shape[0]
    P = np.eye(n+1)
    P[0][0] = 0
    return matrix(P)

def q_(lambda_):
    q = np.vstack((1, -lambda_))
    return matrix(q)

def G_(subgrad_tab):
    k    = len(subgrad_tab)
    ones = np.ones((k, 1))
    tab  = map(np.transpose, subgrad_tab)
    tab  = np.vstack(list(tab))
    
    G = np.hstack((-ones, tab))
    return matrix(G)

def h_(lambda_tab, theta_tab, subgrad_tab):
    k = len(lambda_tab)
    h = np.zeros((k, 1))
    
    for j in range(k):
        h[j] = -theta_tab[j] + np.dot(lambda_tab[j].T, subgrad_tab[j]) 
        
    return matrix(h)
```

#### Itération `next_x`

```python
def bundle_next_x(x, x_tab, f_tab, subgrad_tab):
    P = P_(x)
    q = q_(x)
    G = G_(subgrad_tab[:-1])
    h = h_(x_tab, f_tab[:-1], subgrad_tab[:-1])
    solution = solvers.qp(P, q, G, h)
    x = np.array(solution['x'])[1:]
    return x
```

## Solution du problème primal
---

Pour une matrice $A$ donnée, la fonction que l'on cherche à minimiser est la suivante :

```python
def cost(U, A):
    """
    Total cost of the affectation
    """
    return int(np.trace(np.dot(A.T, U)))
```

La résolution du problème se fait en suivant les étapes suivantes :

1. On dualise le problème et on cré un simulateur `simdual` de la fonction dual $\theta$.

1. En utilisant la fonction `minimize`, on approche le $\lambda$ minimisant $\theta(\lambda)$. 

1. La matrice $U_{\lambda}$ minimisant le coût total de l'affectiation peut alors être calculé avec `U_(lambda_, A)`.

1. On retrouve finalement le coût minimal (approché) avec `cost(U_lambda, A)`.

## Exemples de problèmes d'affectation
---

Pour tester plus facilement le problème d'affectation on cré une fonction `affect` qui prend en entrée une matrice de coûts `A` et qui retournes les solutions `U` et coûts totaux correspondants trouvés par descente de sous-gradient et algorithmes de faisceaux.

```python
np.random.seed(42) # for reproductibility

def affect(A, nb_iter=100, until_feasable=False):
    """
    Solves the assignement problem defined by the matrix A.
    Plots for both the subgradient and the bundle algorithm:
    - the best affectation found
    - theta(lambda_k) along the iterations
    
    Parameters:
    - 'nb_iter' is the number of iterations for the minimization algorithm
    - if 'until_feasable' is True these algorithms will be run multiple
      each time with a different initial value of lambda, starting with lamda_ = [1, ..., 1]
      until the solution found is feasable.
    """
    
    # Simulation of the dual function associated to the affectation problem
    simul_dual = lambda x : simdual(x, A)
    
    # Algorithms that will be used
    algorithms = ['subgradient descent', 'bundle']
    next_xs    = [subgrad_descent_next_x, bundle_next_x]
    
    for algorithm, next_x in zip(algorithms, next_xs):
        
        print(BOLD + 'Launching the {} algorithm...'.format(algorithm) + END)

        # Initial value of lambda = [1, ..., 1]
        lambda_init = np.zeros((A.shape[0], 1)) + 1

        
        with Time() as runtime:
            U = None
            number_of_lambda_init_tested = 0
            
            while not is_permuation_matrix(U):
                lambda_, lambda_tab, theta_tab, subgrad_tab = minimize(f_simulator = simul_dual, 
                                                                       next_x      = next_x,
                                                                       x0          = lambda_init, 
                                                                       nb_iter     = nb_iter)
                # New random value of lambda
                lambda_init = np.random.rand(A.shape[0], 1)*100
                number_of_lambda_init_tested += 1

                U = U_(lambda_, A)
                if not until_feasable: break
        
        # Displaying the best results
        print('Number of initial values of lambda tested:', number_of_lambda_init_tested)
        plot_results(theta_tab, A, U, cost(U, A), runtime())
        
```

```python
def is_permuation_matrix(U):
    """
    Returns True if U is a permutation matrix
    """
    U = np.array(U)
    return (U.ndim == 2 
            and U.shape[0] == U.shape[1] 
            and (U.sum(axis=0) == 1).all() 
            and (U.sum(axis=1) == 1).all() 
            and ((U == 1) | (U == 0)).all())
```

### Problème 1

Commençons par un problème simple dont on peut trouver la solution à la main.

```python
A = np.array([[10, 1, 10], 
              [10, 10, 1], 
              [1, 10, 10]])
```

Ici la meilleur affectation est :
$$
\textrm{U} = 
\begin{bmatrix}
0 & 1 & 0 \\
0 & 0 & 1\\
1 & 0 & 0
\end{bmatrix}
$$

```python
affect(A, nb_iter=5)
```

On retrouve bien la solution espérée.
